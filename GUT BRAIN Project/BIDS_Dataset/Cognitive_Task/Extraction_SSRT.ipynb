```python
import pandas as pd
import numpy as np
```

```python

def ssrt_extract(filename):
    ssrt = pd.read_csv(filename, sep='\t', encoding='UTF-16LE', header= 1)
    
    id_participant = filename[12:15]
    id_participant = int(id_participant)
    
    if (id_participant <=9):
        id_participant = str('RND00')+str(id_participant)
    elif (id_participant >= 10 and id_participant <=95):
        id_participant = str('RND0')+str(id_participant)
    elif (id_participant >= 96 and id_participant <=99):
        id_participant = str('RCX0')+str(id_participant)
    elif (id_participant >= 100 and id_participant <=899):
        id_participant = str('RCX')+str(id_participant)
    elif (id_participant >= 900):
        id_participant = str('RGC')+str(id_participant)
  
    session = filename[11:12]
    if session == '1':
        session = str('1_Baseline')
    elif (session == '2' and id_participant[0:3] == 'RGC'):
          session = str('5_ControlT2')
    elif session == '2':
        session = str('2_4mois')
    elif (session == '3' and id_participant[0:3] == 'RGC'):
          session = str('2_4mois')
    elif session == '3':
        session = str('3_12mois')
    elif session == '4':
        session = str('4_24mois')
    else : session = 'erreur'
        
    Go_Err = ssrt['RES_WRNG'].replace(0,np.NaN).sum()
    p_Go_Err = Go_Err / sum(ssrt['StairCase'].str.count('Go'))
    meanRT_Go_Err = sum(ssrt['RT_WRNG'])/Go_Err
    Go_Success = ssrt['RES_COR'].sum()
    p_Go_Success = Go_Success / sum(ssrt['StairCase'].str.count('Go'))
    Omission_Err = str('en attente')
    p_Omission_Err = str('aussi')
    meanRT_Go_Success = ssrt['RT_COR'].replace(0,np.NaN).mean()
    stdRT_Go_Success = ssrt['RT_COR'].replace(0,np.NaN).std()
    NoGo_Err = ssrt['RES_FA'].sum()
    p_NoGo_Err = NoGo_Err / ssrt['StopTrial'].sum()
    NoGo_Success = ssrt['RES_CNR'].sum()
    p_NoGo_Success = NoGo_Success / ssrt['StopTrial'].sum()
    meanRT_NoGo_Err = ssrt['RT_FA'].replace(0,np.NaN).mean()
    stdRT_NoGo_Err = ssrt['RT_FA'].replace(0,np.NaN).std()
    meanSSD = ssrt['StopDelay'].replace(-1,np.NaN).mean()
    stdSSD = ssrt['StopDelay'].replace(-1,np.NaN).std()
    
    quart1 = ssrt['RT_COR'].replace(0,np.NaN).quantile(.25)
    quart3 = ssrt['RT_COR'].replace(0,np.NaN).quantile(.75)
    iq = (ssrt['RT_COR'].replace(0,np.NaN).quantile(.75) - ssrt['RT_COR'].replace(0,np.NaN).quantile(.25))
    lowOutlier= (quart1-1.5*iq)
    highOutlier= (quart3 + 1.5*iq)
    noOutlier = ssrt[(ssrt['RT_COR'] < highOutlier) & (ssrt['RT_COR'] > lowOutlier)]
    corrected_meanRT_Go_Succes = noOutlier['RT_COR'].mean()
    corrected_stdRT_Go_Succes = noOutlier['RT_COR'].std()
    SSRT = meanRT_Go_Success - meanSSD
    corrected_SSRT = corrected_meanRT_Go_Succes - meanSSD
     
    return bdd_ssrt.append({'id_participant' : id_participant , 'session' : session,  'Go_Err': Go_Err, 
                            'p_Go_Err':p_Go_Err, 'meanRT_Go_Err':meanRT_Go_Err, 'Go_Success':Go_Success, 
                            'p_Go_Success':p_Go_Success,'Omission_Err':Omission_Err, 'p_Omission_Err':p_Omission_Err,
                            'meanRT_Go_Success':meanRT_Go_Success,'stdRT_Go_Success':stdRT_Go_Success, 
                            'NoGo_Err':NoGo_Err, 'p_NoGo_Err':p_NoGo_Err,'NoGo_Success':NoGo_Success,
                            'p_NoGo_Success':p_NoGo_Success,'meanRT_NoGo_Err':meanRT_NoGo_Err,
                            'stdRT_NoGo_Err':stdRT_NoGo_Err,'quart1':quart1, 'quart3':quart3, 'lowOutlier':lowOutlier,
                            'highOutlier':highOutlier,'corrected_meanRT_Go_Succes':corrected_meanRT_Go_Succes,
                            'corrected_stdRT_Go_Succes':corrected_stdRT_Go_Succes, 'meanSSD':meanSSD ,
                            'SSRT':SSRT, 'corrected_SSRT':corrected_SSRT}, ignore_index=True);
      
    
    
    
```

## Netoyage de la BDD

On conserve les patient selon les "lenient criteria" de Congdon et al. : https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3283117/

Our lenient outlier criteria were as follows: 
(1) Percent inhibition on stop trials less than 25% or greater than 75%; 
(2) Percent Go-Response less than than 60%; 
(3) Percent Go-Errors greater than 10%; 
(4) SSRT estimate that is negative or less than 50â€‰ms.

```python
bdd2_ssrt = bdd_ssrt[(bdd_ssrt['p_NoGo_Success'] >= 0.25) & (bdd_ssrt['p_NoGo_Success'] <= 0.75)]
bdd2_ssrt = bdd2_ssrt[bdd2_ssrt['p_Go_Success'] >= 0.6]
bdd2_ssrt = bdd2_ssrt[bdd2_ssrt['p_Go_Err'] <= 0.1]
bdd2_ssrt = bdd2_ssrt[bdd2_ssrt['corrected_SSRT'] >= 50]
```
